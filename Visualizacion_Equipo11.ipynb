# Información gráfica

# Gráficos uni-variantes

### Gráficas de barras

```python
rona.head(20).plot.bar(x='País', y='CasosTotales', color='yellow')
```

Este gráfico nos muestra los contagios totales de los primeros 20 paises en nuestra base de datos (expresado en millones)

```python
rona['Continente'].value_counts().plot.bar(color='cyan')
```

```python
(rona['Continente'].value_counts()/len(rona)).plot.bar(color='cyan')
```

Las gráficas anteriores nos muestran la distribución de terrotorios de acuerdo al continente en que se encuentran de forma cuantitativa y porcentual

### Gráficas de lineas

```python
rona.plot.area(x='Continente', y='CasosCríticos', color='orange')
```

Este gráfico nos muestra como se distribuyen los casos críticos en relación con su continente correspondiente

```python
rona.plot.line(x='Continente', y='MuertesTotales', color='red')
```

La gráfica nos muestra en comportamiento de los fallecimientos por continente

### Histogramas

```python
rona['RecuperadosTotales'].plot.hist(color='green')
```

Este gráfico nos muestra la frecuencia de la cantidad de personas recuperadas (en millones)

```python
rona[rona['CasosCríticos']<100]['CasosCríticos'].plot.hist(color='orange')
```

Este histograma nos muestra la frecuencia de países que cuentas con casos críticos menores a 100

# Gráficos bi-variantes

### Scatter plot

```python
rona.plot.scatter(x='CasosTotales', y='TestTotales', color='blue')
```

```python
rona.plot.hexbin(x='CasosTotales', y='TestTotales', color='blue', gridsize=2)
```

Estos gráficos nos muestran la relación entre las pruebas aplicadas y el total de contagios por región (en millones), podemos decir que en el primero se aprecia mejor su distribución

### Stacked plot

```python
rona.head(5).plot.bar(stacked=True)
```

```python
rona.head(5).plot.line(stacked=True)
```

```python
rona.head(5).plot.area(stacked=True)
```

Este tipo de gráficos no nos sirven de mucho ya que tenemos una alta cantidad de datos a analizar y estos se distorcionan al realizar una gráfica como esta

### Box plot

```python
sns.boxplot(x='Continente', y='MuertesTotales', data=rona)
```

```python
sns.violinplot(x='Continente', y='MuertesTotales', data=rona)
```

Nuestros boxplots no es tan muy bien formados, lo que nos incadica que hay mucha variación entre los datos

### Seaborn

```python
sns.countplot(rona.query('CasosTotales < 100').CasosTotales)
```

La gráfica **Countplot** nos indica la cantidad de regiones que tienen un número de contagios inferior a 100

```python
sns.kdeplot(rona['CasosActivos'], color='orange')
```

El gráfico **KDE** nos indica la densidad de los casos activos (en millones)

```python
sns.jointplot(x='CasosTotales', y='MuertesTotales', data=rona, color='red')
```

El gráfico **Jointplot** nos arroja la relación entre los infectados y los descesos por región

```python
sns.displot(rona.query('MuertesTotales > 10000').MuertesTotales, bins=10,color='red')
```

Este gráfico tipo **Displot** nos muestra la cantidad de regiones con mas de 10000 descesos 

```python
sns.pairplot(rona[['CasosTotales','MuertesTotales','TestTotales']])
```

El gráfico **Pairplot** realiza varias gráficas para comparar distintos datos de nuestra base

# Gráficos multi-variantes

```python
sns.lmplot(x='CasosTotales', y='RecuperadosTotales', hue='Continente', data=rona)
```

Grafico que registra los casos y recuperados totales de acuerdo a su continente

```python
plt.hist(rona[rona['CasosTotales']<1000]['CasosTotales'])
plt.hist(rona[rona['MuertesTotales']<100]['MuertesTotales'])
plt.title("Histograma")
plt.xlabel("Cantidad")
plt.ylabel("Frecuencia")
plt.show()
```

Histogramas de muertes menores a 100 y casos totales menores a 1000 de acuerdo a cuantas regiones cuentan con estos datos

```python
sns.boxplot(x='Continente', y='CasosActivos', hue='Continente', data=rona)
```

Boxplot de los casos activos de acuerdo al continente en que se encuentran

```python
sns.heatmap(rona.corr())
```

Mapa de correlación de la base

# Puntos extra

```python
rona['Continente'].value_counts().plot.pie(x='CasosTotales')
```

```python
(rona['Continente'].value_counts()/len(rona)).plot.bar(color='cyan')
```

Estas dos gráficas nos dan la misma infromación (la dispersión de las regiones de acuerdo al continente en que se encuentran) comparando estas, la gráfica de pie nos muestra de manera mas clara la forma y cantidad de regiones dentro de cada continente ya que se puede ver de una forma mas visual que la gráfica de barras y en nuestra opinion tiene una mejor estetica
